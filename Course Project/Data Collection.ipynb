```python
'''Import necessary library'''
import metacritic
import pickle
import pandas as pd
from math import sqrt
```

```python
'''Get all critics and creates a dictionary where the key is each critic's url
here the critic_url is in the form /critic/tom-meek as written in the critics file'''

all_critics = metacritic.get_all_movie_critics()

critic_dict = {}
for c in all_critics:
    critic_dict[(c['critic_url'])] = c['critic_name'].encode('utf-8') 
```

```python
'''Dataset Collection'''
"""Code used to create reviews file"""

"""Reviews file: consists of the reviews and scores of all critics in critics.txt - training data"""
"""Function Returns dataset of movies and their rating by critic"""
def get_all_reviews():
    #reads file containing the url of each critic (100 critics)
    critics = open("critics.txt", "r")
    #for all critics, write each review and score to a file
    reviews = open("reviews.txt", "w")
    reviews.write(str("Review") + '\t' + str("Sentiment") + '\n')
    for c in critics:
        r = metacritic.get_reviews_by_critic(("http://www.metacritic.com" + c))
        for review in r:
            reviews.write((review['review_body'].encode('utf-8') + '\t' + str(review['score']) + '\n'))
        
    reviews.close()
    critics.close() 
    return
```

```python
'''Method to get a dictionary of critics along with their movie ratings.'''

def get_scores():
    movie_dataset = {}
    critics = open("critics.txt","r")
    for c in critics:
        r = metacritic.get_reviews_by_critic(("http://www.metacritic.com" + c))
        movies = {}
        for review in r:
            movies[review['movie_name'].encode('utf-8')] = review['score']
        movie_dataset[critic_dict[c.strip('\n')]] = movies
        
    with open('dict_scores_for_25_critics.txt', 'wb') as handle:
        pickle.dump(movie_dataset, handle)
        
    return 
```

```python
dataset = get_all_reviews()
scores = get_scores()
```
